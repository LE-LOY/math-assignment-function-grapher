```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.ioff()

from sympy import *
init_printing()
```

```python
t, c = symbols('t c')
a, v, s = symbols('a v s')
```

```python
def integrate_func(f, t, given_point):
    """  
    
    """
    f_i = integrate(f, t)
    c = symbols('c')
    
    temp = f_i + c
    
    res = f_i + solve(temp.subs(t, given_point[0]) - given_point[1], c)[0]
    
    return res
```

```python
def get_integrates(f, t, depth, given_points):
    """  
    
    """
    res = [f]
    for i in range(1, depth + 1):
        func = integrate_func(res[i - 1], t, given_points[i - 1])
        res.append(func)
    return res
```

```python
def plot_funcs(a, v, s):
    """  
    
    """
    p = plot(a, v, s, xlim=(0, 10), ylim=(0, 100), show=False) 
    
    p[0].line_color='r'
    p[1].line_color='g'
    p[2].line_color='b'
    
    p.show()
```

### Problem 1

#### Equations

```python
a, v, s = get_integrates(sympify('1.2*t'),
                         t,
                         depth=2,
                         given_points=[(1, 5), (1, 6)])
```

```python
a
```

```python
v
```

```python
s
```

#### Plot

```python
plot_funcs(a, v, s)
```

#### Answers

Velocity at time = 2:

```python
v.subs(t, 2)
```

Displacement at time = 2:

```python
s.subs(t, 2)
```

#### Answers

Time at maximum velocity:

```python
t_max = solve(a, t)
t_max
```
